```python
Genuine_sample = Genuine.sample(n=492)
```

# Concatenating the 2 Data frames

```python
Sample_data =pd.concat([Genuine_sample,Fraud],axis=0)
Sample_data.head()
```

# MODEL

# Logistic Regression

Splitting Data into Features & Targets

```python
x = Sample_data.drop('Class', axis='columns')
y = Sample_data['Class']
```

Splitting data into training & testing

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=2,stratify=y)
```

```python
cols_to_scale = ['Time','Amount']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

Sample_data[cols_to_scale] = scaler.fit_transform(Sample_data[cols_to_scale])
```

```python
model = LogisticRegression()
```

```python
#train the model
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
print(predictions)
```

```python
print(classification_report(y_test, predictions))
```

The accuracy of the model is 92%

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=200, random_state=5)
classifier.fit(x_train, y_train)
predictions = classifier.predict(x_test)
```

```python
feat_importances = pd.Series(classifier.feature_importances_, index=x.columns)
feat_importances.nlargest(5).plot(kind='barh')
```

From the above graph, it shows that, V14, V10, V12, V11 and V17 were the highest determinants of the fraud.
